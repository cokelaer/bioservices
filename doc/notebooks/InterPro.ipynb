# Accessing [InterPro](https://www.ebi.ac.uk/interpro/) with BioServices

This notebook illustrates some of the functionalities of the InterPro service accessible from BioServices.

InterPro provides functional analysis of proteins by classifying them into families and predicting domains and important sites. It integrates signatures from multiple member databases (Pfam, PRINTS, ProSite, SMART, etc.) into a single searchable resource.

## <a name="setup"></a> Setup

```python
from bioservices import InterPro
i = InterPro()
```

## <a name="entries"></a> Entry queries

InterPro entries group together proteins with similar sequence and function. Each entry is identified by an accession of the form `IPRxxxxxx`.

### Retrieve a specific InterPro entry

```python
entry = i.get_entry("IPR000001")
entry["metadata"]["name"]
```

```python
print("Accession:", entry["metadata"]["accession"])
print("Name     :", entry["metadata"]["name"])
print("Type     :", entry["metadata"]["type"])
```

### Browse all InterPro entries (paginated)

```python
results = i.get_entries(page_size=5, page=1)
print("Total entries:", results["count"])
for r in results["results"]:
    print(r["metadata"]["accession"], "-", r["metadata"]["name"])
```

### Search entries by keyword

```python
results = i.search_entries("kinase", page_size=5)
print("Matches for 'kinase':", results["count"])
for r in results["results"]:
    print(r["metadata"]["accession"], "-", r["metadata"]["name"])
```

### Filter entries by type

Entries can be filtered by type: `family`, `domain`, `homologous_superfamily`, `repeat`, `site`, `active_site`, `binding_site`, `conserved_site`, or `ptm`.

```python
results = i.get_entries_by_type("domain", page_size=5)
print("Total domain entries:", results["count"])
for r in results["results"]:
    print(r["metadata"]["accession"], "-", r["metadata"]["name"])
```

### Retrieve entries from a member database

InterPro integrates entries from many member databases including Pfam, ProSite, SMART, CDD, HAMAP, etc.

```python
# Get a specific Pfam entry
pfam_entry = i.get_member_database_entry("pfam", "PF00001")
print("Pfam entry name:", pfam_entry["metadata"]["name"])
```

```python
# Browse all entries from the Pfam database
pfam_results = i.get_entries_by_member_database("pfam", page_size=5)
print("Total Pfam entries:", pfam_results["count"])
for r in pfam_results["results"]:
    print(r["metadata"]["accession"], "-", r["metadata"]["name"])
```

## <a name="proteins"></a> Protein queries

You can look up proteins by their UniProt accession and retrieve the InterPro entries that annotate them.

### Retrieve information about a protein

Here we look up human prothrombin (coagulation factor II), UniProt accession P00734.

```python
protein = i.get_protein("P00734")
print("Protein name:", protein["metadata"]["name"])
print("Source      :", protein["metadata"]["source_database"])
print("Length      :", protein["metadata"]["length"])
```

### Get InterPro entries annotated on a protein

```python
entries = i.get_protein_entries("P00734")
print("InterPro entries for P00734:")
for r in entries["results"]:
    print(r["metadata"]["accession"], "-", r["metadata"]["name"])
```

### Get proteins annotated with a specific InterPro entry

```python
proteins = i.get_proteins_by_entry("IPR000001", page_size=5)
print("Total proteins for IPR000001:", proteins["count"])
for r in proteins["results"]:
    print(r["metadata"]["accession"], "-", r["metadata"]["name"])
```

## <a name="structures"></a> Structure queries

InterPro links entries to 3D structures deposited in the Protein Data Bank (PDB).

### Retrieve information about a PDB structure

```python
structure = i.get_structure("1t2v")
print("PDB ID  :", structure["metadata"]["accession"])
print("Name    :", structure["metadata"]["name"])
```

### Get PDB structures associated with an InterPro entry

```python
structures = i.get_entry_structures("IPR000001", page_size=5)
print("Total structures for IPR000001:", structures["count"])
for r in structures["results"]:
    print(r["metadata"]["accession"], "-", r["metadata"]["name"])
```

## <a name="taxonomy"></a> Taxonomy queries

Taxonomy information allows you to explore the distribution of InterPro entries across species.

### Retrieve taxonomy information for a taxon

```python
# Taxonomy ID 9606 corresponds to Homo sapiens
taxon = i.get_taxonomy("9606")
print("Scientific name:", taxon["metadata"]["name"]["name"])
print("Full name      :", taxon["metadata"]["name"]["short"])
```

### Get taxonomy distribution for an InterPro entry

```python
taxons = i.get_entry_taxonomy("IPR000001", page_size=5)
print("Total taxa for IPR000001:", taxons["count"])
for r in taxons["results"]:
    print(r["metadata"]["accession"], "-", r["metadata"]["name"])
```

## <a name="proteomes"></a> Proteome queries

Proteomes represent the complete set of proteins expressed by an organism.

### Retrieve information about a proteome

UP000005640 is the reference proteome for *Homo sapiens*.

```python
proteome = i.get_proteome("UP000005640")
print("Proteome accession:", proteome["metadata"]["accession"])
print("Name              :", proteome["metadata"]["name"])
print("Is reference      :", proteome["metadata"]["is_reference"])
```

### Get proteomes containing proteins annotated with an InterPro entry

```python
proteomes = i.get_entry_proteomes("IPR000001", page_size=5)
print("Total proteomes for IPR000001:", proteomes["count"])
for r in proteomes["results"]:
    print(r["metadata"]["accession"], "-", r["metadata"]["name"])
```

## <a name="sets"></a> Set queries

Sets represent groups of related entries within a member database, such as Pfam clans.

### Retrieve a Pfam clan

```python
pfam_clan = i.get_set("pfam", "CL0001")
print("Clan accession:", pfam_clan["metadata"]["accession"])
print("Clan name     :", pfam_clan["metadata"]["name"])
```
