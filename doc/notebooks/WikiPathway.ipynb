# Accessing [WikiPathways](https://www.wikipathways.org/) with BioServices

**WikiPathways** is an open, collaborative platform for capturing and disseminating
models of biological pathways for data visualisation and analysis.

This notebook shows how to:

- Find pathways by keyword or gene
- Retrieve pathway metadata
- Display pathway diagrams

> **WikiPathways REST API**: https://webservice.wikipathways.org/


```python
import matplotlib.pyplot as plt
import pandas as pd
from bioservices import WikiPathways

s = WikiPathways()
# Default organism is Homo sapiens
print('Default organism:', s.organism)
```

## 1. Finding pathways by keyword

Use `findPathwaysByText` to search all WikiPathways for a keyword.


```python
# Find pathways related to BRAF
res = s.findPathwaysByText("BRAF")
print(f"Found {len(res)} pathway(s) related to BRAF")
# Show human pathways
human = res.query("species == 'Homo sapiens'")
print(f"  Human pathways: {len(human)}")
for id_, row in human.iterrows():
    print(f"  {id_}: {row['species']}")
```

```python
# Find pathways related to MTOR signalling
mtor = s.findPathwaysByText("MTOR")
human_mtor = mtor.query("species == 'Homo sapiens'")
print(f"Found {len(human_mtor)} human MTOR pathways")
for id_,p in human_mtor.iterrows():
    print(f"  {id_}: {p['name']}")
```

## 2. Retrieving pathway information

Use `getPathwayInfo` to get metadata for a specific pathway identified by
its WikiPathways identifier (e.g. `WP734`).


```python
# Retrieve info for the SARS-CoV-2 / COVID-19 pathway
info = s.getPathwayInfo("WP734")
print(type(info))
if hasattr(info, '__dict__'):
    print(vars(info))
elif isinstance(info, dict):
    for k, v in info.items():
        print(f'  {k}: {v}')
else:
    print(info)
```

## 3. Listing pathways for an organism

Use `listPathways` to retrieve all pathways available for a given species.


```python
# List all human pathways (may take a few seconds)
all_human = s.listPathways("Homo sapiens")
print(f"Total human pathways: {len(all_human)}")
```

## 4. Displaying a pathway diagram

Use `getColoredPathway` to retrieve an SVG of the pathway with optional
node colouring, then display it in the notebook with `IPython.display.SVG`.


```python
from IPython.display import SVG, display

# Retrieve the SVG for the B-cell receptor signalling pathway (WP23)
svg_data = s.getColoredPathway("WP23")
if svg_data and str(svg_data).startswith('<'):
    display(SVG(svg_data))
else:
    print('SVG not available:', str(svg_data)[:100])
```

## 5. Finding interactions involving a gene

Use `findInteractions` to list all documented interactions for a given gene
or protein symbol.


```python
# Find interactions involving SMAD2
interactions = s.findInteractions("SMAD2")
print(f"Found {len(interactions)} interactions involving SMAD2")
for inter in interactions[:5]:
    print(' ', inter)
```
